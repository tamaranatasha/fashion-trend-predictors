```python
#Google Trends collcetion script
pytrends = TrendReq(hl='en-GB', tz=0)

trend_list = pd.read_csv("trends_list.csv")

all_data = []

for trend in trend_list["trend_name"]:

    try:
        pytrends.build_payload(
            [trend],
            timeframe='2021-01-01 2025-12-31',
            geo=''
        )

        df = pytrends.interest_over_time()

        if not df.empty:

            df = df.reset_index()

            df["trend_name"] = trend

            df = df.rename(
                columns={trend: "google_score"}
            )

            all_data.append(
                df[["date",
                    "trend_name",
                    "google_score"]]
            )

        print(f"Completed: {trend}")

        time.sleep(3)

    except Exception as e:

        print(
            f"Error with {trend}: {e}"
        )

combined = pd.concat(all_data)

combined.to_csv(
    "fashion_trends_raw.csv",
    index=False
)

print("Dataset saved.")
```

```python
#import csv
df = pd.read_csv("fashion_trends_raw.csv")
categories = pd.read_csv("trends_list.csv")
#merge dfs
df = df.merge(categories,
              on="trend_name",
              how="left")
df.to_csv("fashion_trends_combined.csv",
         index=False)
df = pd.read_csv("fashion_trends_combined.csv")
```
